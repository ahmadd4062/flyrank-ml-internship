## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## Refresh / Content Opportunity Scoring

My lane is primarily a **ranking** task. The goal is to rank content pages based on their likelihood of benefiting from a content refresh. Instead of predicting a single correct answer, the model prioritizes pages so that content teams can review the highest-value opportunities first. A ranking approach is more appropriate because multiple pages may deserve attention, but some should be reviewed before others.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The ideal target is whether a content page is a good candidate for refresh. Since there is no direct label indicating this, a proxy target can be created using observed performance signals such as declining traffic, declining impressions, or sustained downward trends.

The target is therefore based on observed outcomes in the data rather than a manually assigned label. It serves as a decision-support signal for prioritizing content review.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric is **NDCG (Normalized Discounted Cumulative Gain)** because the project focuses on ranking pages rather than making a simple yes/no prediction.

A good model should rank the most valuable refresh opportunities near the top of the list so that reviewers spend their time on the pages with the greatest potential impact.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

The features describe characteristics of that page, such as search impressions, sessions, click-through rate, trend direction, and other observed search signals.

The output for each row is a refresh priority score or ranking that indicates how strongly the page should be considered for review.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule such as "refresh every page with declining traffic" is unlikely to perform well because content performance depends on many interacting factors. Search impressions, sessions, click-through rate, trend direction, content freshness, and other signals all contribute to whether a page should be reviewed.

Machine learning can learn patterns from multiple features simultaneously and produce a ranked list of pages based on observed relationships in the data. This provides better decision-support than relying on a single threshold or manually written rule.
