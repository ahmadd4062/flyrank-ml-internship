## 1. Unit of analysis + time window

*One row = one what, over which dates? State it, then verify it below.*

## Unit of Analysis + Time Window

- **One row** represents the performance of one content page (`content_hash_id`) for one client (`client_hash_id`) on one reporting date (`report_date`).
- **Time window:** March 2026 (`month=2026-03`) from the `fact_content_daily_performance` table.
- This time window is used because it is a mid-panel month and avoids using the final month as development data.

## 2. Fields: feature / label / context / excluded

*Sort every field you plan to touch into these four buckets. Excluded needs a why.*

## Fields

### Features
- gsc_impressions
- gsc_clicks
- gsc_avg_position
- client_has_gsc
- gsc_data_available

These features are known at the decision time because they are historical search performance signals available before deciding whether a page should be reviewed.

### Label / Proxy
Refresh opportunity priority (proxy).

There is no direct "needs refresh" label in the dataset, so the model uses observed search performance as a proxy for prioritizing pages for review.

### Context
- report_date
- client_hash_id
- content_hash_id

These identify when the observation occurred and which client and content page it belongs to.

### Excluded
GA4 metrics (pageviews, sessions, users, engagement metrics) are excluded because they are unavailable for many clients, resulting in many missing values. AI referral fields are also excluded because they are not consistently available across all observations.

## 3. Verify it with queries (grain, counts, missing values, windows)

*Every claim above gets a query cell here. A contract claim without a query next to it is a guess.*


```python
# Query 1: Grain (What does one row represent)
print("Total rows:", len(df))

print("\nUnique report dates:", df["report_date"].nunique())
print("Unique content pages:", df["content_hash_id"].nunique())
print("Unique clients:", df["client_hash_id"].nunique())

duplicates = df.duplicated(
    subset=["report_date", "client_hash_id", "content_hash_id"]
).sum()

print("\nDuplicate (date, client, content) rows:", duplicates)
```

The query confirms the grain of the dataset. There are no duplicate combinations of report date, client, and content page, supporting the claim that one row represents one content page for one client on one reporting date.

```python
# Query 2: Row count + date span
print("Rows:", len(df))
print("Start date:", df["report_date"].min())
print("End date:", df["report_date"].max())
```

The selected partition contains 9,841,378 rows covering the period from 2026-03-01 to 2026-03-31.

```python
# Query 3: Availability (IS TRUE equivalent)
available = df[
    (df["gsc_data_available"] == True)
]

print("Rows with GSC data available:", len(available))
```

Filtering on `gsc_data_available == True` leaves 3,611,061 rows. This verifies that not every observation contains Search Console data, so feature availability must be considered during analysis.

## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

## Data Limits

This dataset contains historical search-performance observations rather than direct labels indicating whether refreshing a page improved future performance. Therefore, the model provides decision support rather than proof that a refresh will increase traffic or rankings.

Some fields, particularly GA4 metrics, are unavailable for clients without Google Analytics integration, resulting in missing values. Additionally, this analysis uses a single monthly partition (March 2026), so the observations represent that time window only and should not be interpreted as long-term trends.
